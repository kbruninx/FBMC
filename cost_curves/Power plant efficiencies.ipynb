```python
import pandas as pd
import numpy as np
```

```python
zones =  ["ALBE", "ALDE", "AT", "BE", "CZ", "DE_LU", "FR", "HR", "HU", "NL", "PL", "RO", "SI", "SK"]

plant_types = [
    "biomass", "brown_coal", "coal_gas", "natural_gas", "hard_coal", "oil", "hydro", 
    "nuclear", "waste", "other"
] 
```

```python
df_kyos = pd.read_excel('./kyos_plant_database.xlsx')
display(df_kyos)
```

```python
efficiencies = np.zeros((len(zones), len(plant_types)))

natural_gas_i = plant_types.index('natural_gas')
brown_coal_i = plant_types.index('brown_coal')
hard_coal_i = plant_types.index('hard_coal')
oil_i = plant_types.index('oil')

for zone_i, zone in enumerate(zones):
    if zone == 'DE_LU':
        df_kyos_z = df_kyos[df_kyos.powercommid == 'DE']
    else:
        df_kyos_z = df_kyos[df_kyos.powercommid == zone]
    
    natural_gas_eff = round(((df_kyos_z[df_kyos_z.reportingcommid == 'natural_gas'].effmax + df_kyos_z[df_kyos_z.reportingcommid == 'natural_gas'].effmin)/2).mean(), 2)
    brown_coal_eff = round(((df_kyos_z[df_kyos_z.reportingcommid == 'brown_coal'].effmax + df_kyos_z[df_kyos_z.reportingcommid == 'brown_coal'].effmin)/2).mean(), 2)
    hard_coal_eff = round(((df_kyos_z[df_kyos_z.reportingcommid == 'hard_coal'].effmax + df_kyos_z[df_kyos_z.reportingcommid == 'hard_coal'].effmin)/2).mean(), 2)
    oil_eff = round(((df_kyos_z[df_kyos_z.reportingcommid == 'oil'].effmax + df_kyos_z[df_kyos_z.reportingcommid == 'oil'].effmin)/2).mean(), 2)
    
    if np.isnan(natural_gas_eff):
        natural_gas_eff = round(((df_kyos[df_kyos.reportingcommid == 'natural_gas'].effmax + df_kyos[df_kyos.reportingcommid == 'natural_gas'].effmin)/2).mean(), 2)
    if np.isnan(brown_coal_eff):
        brown_coal_eff = round(((df_kyos[df_kyos.reportingcommid == 'brown_coal'].effmax + df_kyos[df_kyos.reportingcommid == 'brown_coal'].effmin)/2).mean(), 2)
    if np.isnan(hard_coal_eff):
        hard_coal_eff = round(((df_kyos[df_kyos.reportingcommid == 'hard_coal'].effmax + df_kyos[df_kyos.reportingcommid == 'hard_coal'].effmin)/2).mean(), 2)
    if np.isnan(oil_eff):
        oil_eff = round(((df_kyos[df_kyos.reportingcommid == 'oil'].effmax + df_kyos[df_kyos.reportingcommid == 'oil'].effmin)/2).mean(), 2)
    
    efficiencies[zone_i, natural_gas_i] = natural_gas_eff
    efficiencies[zone_i, brown_coal_i] = hard_coal_eff
    efficiencies[zone_i, hard_coal_i] = brown_coal_eff
    efficiencies[zone_i, oil_i] = oil_eff
    
```

```python
efficiencies
```

```python
with open('plant_efficiencies.npy', 'wb') as f:
    np.save(f, efficiencies)
```
