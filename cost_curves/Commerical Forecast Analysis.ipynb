```python
forecast_zone_list = []

for zone in countries:
    print(zone)
    df_forecast_zone = pd.DataFrame()
    curve = session.get_curve(name='pri {0} spot ec00 €/mwh cet h f'.format(zone.lower()))
    start_date = pd.Timestamp('2023-1-31 08:00')
    for day in range(28):
        current_date = start_date + pd.offsets.Day(day)
        ts = curve.get_instance(issue_date=current_date)
        df = ts.to_pandas()
        df_forecast_zone = pd.concat([df_forecast_zone, df[0:24]])
    forecast_zone_list.append(df_forecast_zone)
```

```python
df_forecast = forecast_zone_list[0].copy()
for index, zone in enumerate(countries):
    df_forecast[zone] = forecast_zone_list[index][0]

df_forecast = df_forecast.drop(columns=[0])
df_forecast.index = df_forecast.index.tz_localize(None)
df_forecast.to_excel("./price_forecasts/volue_day_ahead_ec00.xlsx")
display(df_forecast)
```
