```python
import pandas as pd
import numpy as np

countries = ['AT', 'BE', 'CZ', 'DE_LU', 'FR', 'HR', 'HU', 'NL', 'PL', 'RO', 'SI', 'SK']

reduced_gen_types = [
    'biomass', 'brown_coal', 'coal_gas', 'natural_gas', 'hard_coal', 'oil', 'hydro', 
    'nuclear', 'waste', 'other', 'solar', 'wind_onshore', 'wind_offshore', 'other_renewable'
]
```

```python
df_gen_single = pd.DataFrame()

for index, country_code in enumerate(countries):
    df_gen = pd.read_excel("./optimisation/data/generation.xlsx", sheet_name=country_code, index_col=0)
    if index == 0:
        df_gen_single.index = df_gen.index
    for gen_type in reduced_gen_types:
        df_gen_single[country_code + '_' + gen_type] = df_gen[gen_type]
        
display(df_gen_single)
```

```python
df_gen_single.to_csv('./optimisation/representative_periods/generation.csv', delimiter=';')
```
