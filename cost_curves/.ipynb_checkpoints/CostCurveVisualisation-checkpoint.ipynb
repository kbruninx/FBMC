```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

countries = ['AT', 'BE', 'CZ', 'DE_LU', 'FR', 'HR', 'HU', 'NL', 'PL', 'RO', 'SI', 'SK']
```

```python
df_coeffs_nl = pd.read_excel("./optimisation/cost_coefficients_wo_np.xlsx", sheet_name='NL')
df_capacities = pd.read_excel("./optimisation/data/installed_capacities.xlsx")
df_fuel_prices = pd.read_excel("./optimisation/data/validation/fuel_prices.xlsx")


nl_gas_max = df_capacities.iloc[3]['NL']
df_coeffs_nl_gas = df_coeffs_nl.iloc[3]

print(df_coeffs_nl_gas)

def c(x):
    return df_coeffs_nl_gas.alpha + (df_coeffs_nl_gas.beta * x + df_coeffs_nl_gas.gamma * x**2)

x = np.linspace(0, nl_gas_max, 100)

plt.plot(x, c(x))
```
