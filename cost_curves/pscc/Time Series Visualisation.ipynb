```python
import pandas as pd
from matplotlib import pyplot as plt
from IPython.display import display, Math
import numpy as np

countries = ['AT', 'BE', 'CZ', 'DE_LU', 'FR', 'HR', 'HU', 'NL', 'PL', 'RO', 'SI', 'SK']

forecast_legend = [
    r'$\mathrm{IO_{L1}}$',
]

start_date = pd.Timestamp('20230201')
end_date = pd.Timestamp('20230728')
```

```python
df_observed = pd.read_excel("../data-july/day_ahead_prices.xlsx", index_col=0)
df_observed = df_observed.loc[(df_observed.index > start_date) & (df_observed.index <= end_date)]

df_forecasts = []

df_forecasts.append(pd.read_excel("../price_forecasts/price_forecast_until_july_2_february-july.xlsx"))

for index, df in enumerate(df_forecasts):
    df_forecasts[index].index = df_observed.index
```

```python
start_date_show = pd.Timestamp('20230201')
end_date_show = pd.Timestamp('20230228')

for zone in countries:
    plt.figure(figsize=(15,6))
    ax = df_observed.loc[(df_observed.index > start_date_show) & (df_observed.index <= end_date_show)][zone].plot()
    for df in df_forecasts:
        df.loc[(df.index > start_date_show) & (df.index <= end_date_show)][zone].plot(ax=ax)
    
    plt.title("{0} day-ahead market prices between April and July, 2023".format(zone))
    plt.ylabel('Price [/MWh]')
    plt.xlabel('Time [day] (hourly resolution)')
    ax.legend(['observed'] + forecast_legend)
    plt.show()
```
