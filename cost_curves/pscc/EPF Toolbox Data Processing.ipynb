```python
import pandas as pd
```

```python
countries = ['AT', 'BE', 'CZ', 'DE_LU', 'FR', 'HR', 'HU', 'NL', 'PL', 'RO', 'SI', 'SK']

borders = {
    'AT': ['DE_LU', 'CZ', 'HU', 'SI'], 
    'BE': ['FR', 'NL', 'DE_LU'], 
    'CZ': ['DE_LU', 'PL', 'SK', 'AT'], 
    'DE_LU': ['NL', 'PL', 'BE', 'FR', 'CZ', 'AT'], 
    'FR': ['BE', 'FR'], 
    'HR': ['HU', 'SI'], 
    'HU': ['AT', 'SK', 'SI', 'RO', 'HR'], 
    'NL': ['BE', 'DE_LU'], 
    'PL': ['DE_LU', 'CZ', 'SK'], 
    'RO': ['HU'], 
    'SI': ['AT', 'HU', 'HR'], 
    'SK': ['PL', 'CZ', 'HU']
}
```

```python
df_da_prices = pd.read_excel("../data-benchmark/day_ahead_prices.xlsx", index_col=0)
df_demand = pd.read_excel("../data-benchmark/demand.xlsx", index_col=0)
df_ren_gen = pd.read_excel("../data-benchmark/renewable_generation.xlsx", index_col=0)
df_gen_forecast = pd.read_excel("../data-benchmark/generation_forecast.xlsx", index_col=0)

df_da_prices = df_da_prices.fillna(0)
df_demand = df_demand.fillna(0)
df_ren_gen = df_ren_gen.fillna(0)
df_gen_forecast = df_gen_forecast.fillna(0)
```

```python
# renewable generation as exogenous

for zone in countries:
    print(zone)
    df_dataset = pd.DataFrame()
    df_dataset['Date'] = df_da_prices.index
    df_dataset.set_index('Date', inplace = True)
    df_dataset['Price'] = df_da_prices[zone]
    df_dataset['Exogenous 1'] = df_demand[zone]
    #df_dataset['Exogenous 2'] = df_ren_gen[zone]
    df_dataset['Exogenous 2'] = df_gen_forecast[2:][zone]

    #df_dataset['Exogenous 4'] = 0
    #df_dataset['Exogenous 5'] = 0
    #df_dataset['Exogenous 6'] = 0
    
    #if len(borders[zone]) > 0:
    #    df_dataset['Exogenous 4'] = df_gen_forecast[2:][borders[zone][0]]
    #if len(borders[zone]) > 1:
    #    df_dataset['Exogenous 5'] = df_gen_forecast[2:][borders[zone][1]]
    #if len(borders[zone]) > 2:
    #    df_dataset['Exogenous 6'] = df_gen_forecast[2:][borders[zone][2]]

    df_dataset = df_dataset.fillna(0)
    df_dataset.to_csv('./datasets/{0}_CORE.csv'.format(zone))
```
