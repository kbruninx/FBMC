```python
import pandas as pd
from matplotlib import pyplot as plt

countries = ['AT', 'BE', 'CZ', 'DE_LU', 'FR', 'HR', 'HU', 'NL', 'PL', 'RO', 'SI', 'SK']
```

```python
df_observed = pd.read_excel("./optimisation/data/day_ahead_prices.xlsx", index_col=0)
df_observed = df_observed.drop(columns=['ALBE', 'ALDE'])
df_observed = df_observed[1466:1466+5*24]

df_forecast1 = pd.read_excel("./optimisation/price_forecast_w_outages.xlsx")
df_forecast1 = df_forecast1[1:1+5*24]
df_forecast1.index = df_observed.index

df_forecast2 = pd.read_excel("./optimisation/price_forecast_rolling_test.xlsx")
df_forecast2.index = df_observed.index
```

```python
for zone in countries:
    plt.figure(figsize=(15,6))
    ax = df_observed[zone].plot()
    df_forecast1[zone].plot(ax=ax)
    df_forecast2[zone].plot(ax=ax)
    plt.title("{0} day-ahead market prices in November, 2022 (w/o NP obs)".format(zone))
    plt.ylabel('Price [€]')
    plt.xlabel('Time [day] (hourly resolution)')
    ax.legend(['observed', 'forecast global', 'rolling'])
    plt.show()
```
