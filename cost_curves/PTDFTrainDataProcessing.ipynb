```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import numpy as np

hubs = ['ALBE', 'ALDE', 'AT', 'BE', 'CZ', 'DE_LU', 'FR', 'HR', 'HU', 'NL', 'PL', 'RO', 'SI', 'SK']
```

```python
df_ptdf = pd.read_excel("./for_io/training/ptdfs.xlsx", index_col=0)
```

```python
max_cne_num = np.max(df_ptdf.groupby(df_ptdf.index).eic.count())
df_ptdf_empty_list = []

for time in df_ptdf.index.unique():
    print(time)
    cne_num = df_ptdf.loc[time].eic.count()
    dummy_cne_to_add = max_cne_num - cne_num

    columns = ['eic', 'ram'] + hubs
    empty_content = [[None for i in range(16)] for j in range(dummy_cne_to_add)]
    df_ptdf_empty_list.append(pd.DataFrame(empty_content, columns=columns, index=[time for i in range(dummy_cne_to_add)]))
```

```python
df_ptdf = pd.concat([df_ptdf] + df_ptdf_empty_list)
df_ptdf = df_ptdf.sort_index()
```

```python
df_ptdf.to_excel("./optimisation/data/training/ptdfs.xlsx")
```
