```python
start = pd.Timestamp('20220901', tz='Europe/Amsterdam')
end = pd.Timestamp('20220902', tz='Europe/Amsterdam')
#end = pd.Timestamp('20230228', tz='Europe/Amsterdam')

countries = ['AT', 'BE', 'CZ', 'DE_LU', 'FR', 'HR', 'HU', 'NL', 'PL', 'RO', 'SI', 'SK']

gen_types = [
    'Biomass', 'Fossil Brown coal/Lignite', 'Fossil Coal-derived gas', 'Fossil Gas', 'Fossil Hard coal', 
    'Fossil Oil', 'Fossil Oil shale', 'Fossil Peat', 'Geothermal', 'Hydro Pumped Storage', 
    'Hydro Run-of-river and poundage', 'Hydro Water Reservoir', 'Marine', 'Nuclear', 'Other',
    'Other renewable', 'Solar', 'Waste', 'Wind Offshore', 'Wind Onshore'
]

standardized_gen_types = [
    'biomass', 'brown_coal', 'coal_gas', 'natural_gas', 'hard_coal', 'oil', 'oil_shale', 'peat', 'geothermal', 
    'hydro_pumped_storage', 'hydro_run_of_river', 'hydro_reservoir', 'marine', 'nuclear', 'other', 'other_renewable',
    'solar', 'waste', 'wind_offshore', 'wind_onshore'
]

ppm_types = ['Nuclear', 'Hard Coal', 'Natural Gas', 'Hydro', 'Lignite', 'Oil', 'Bioenergy', 'Other', 'Wind', 'Solar', 'Waste']
ppm_type_map = ['nuclear', 'hard_coal', 'natural_gas', 'hydro', 'brown_coal', 'oil', 'biomass', 'other', 'wind', 'solar', 'waste']
```

```python
df_powerplants_db = pd.read_excel('./power_plants.xlsx', index_col=0)
df_powerplants_db.eic = df_powerplants_db.eic.str.replace("{'", "")
df_powerplants_db.eic = df_powerplants_db.eic.str.replace("'}", "")
df_powerplants_db.eic = df_powerplants_db.eic.astype(str)

for index, row in df_powerplants_db.iterrows():
    df_powerplants_db.loc[index, 'type'] = ppm_type_map[ppm_types.index(row['fuel_type'])]

display(df_powerplants_db)
```

```python
df_plants_entsoe = pd.DataFrame()

for (index, country_code) in enumerate([x for x in countries if x not in ['HR']]):
    print(country_code)
    df_installed = client.query_installed_generation_capacity_per_unit(country_code, start=start, end=end)
    if index == 0:
        df_plants_entsoe['name'] = df_installed['Name']
        df_plants_entsoe['eic'] = df_installed.index
        df_plants_entsoe['type'] = df_installed['Production Type']
        df_plants_entsoe['zone'] = country_code
        df_plants_entsoe['installed_capacity'] = df_installed['Installed Capacity [MW]']
    else:
        df_plants_entsoe_temp = pd.DataFrame()
        df_plants_entsoe_temp['name'] = df_installed['Name']
        df_plants_entsoe_temp['eic'] = df_installed.index
        df_plants_entsoe_temp['type'] = df_installed['Production Type']
        df_plants_entsoe_temp['zone'] = country_code
        df_plants_entsoe_temp['installed_capacity'] = df_installed['Installed Capacity [MW]']
        df_plants_entsoe = pd.concat([df_plants_entsoe, df_plants_entsoe_temp], ignore_index=True)
```

```python
for index, plant in df_plants_entsoe.iterrows():
    df_plants_entsoe.loc[index, 'type'] = standardized_gen_types[gen_types.index(plant['type'])]
    plants_location = df_powerplants_db[df_powerplants_db.eic == plant['eic']]
    if len(plants_location) > 0:
        df_plants_entsoe.loc[index, 'lat'] = plants_location.iloc[0]['lat']
        df_plants_entsoe.loc[index, 'lon'] = plants_location.iloc[0]['lon']
    
df_plants_entsoe = df_plants_entsoe[~df_plants_entsoe.type.isin(['other_renewable', 'solar', 'wind_offshore', 'wind_onshore'])]
```
